## 1 ลบทิ้ง ```dropna()```

````python
import pandas as pd
````

````python
SS_warwick.head()
````

### Drop เฉพาะ column ที่มี missing มากกว่า 50%

````python
# Calculate the percentage of missing values for each column
missing_percentage = SS_warwick.isnull().sum() / len(SS_warwick) * 100

# Identify columns with more than 50% missing values
columns_to_drop = missing_percentage[missing_percentage > 50].index

# Drop the identified columns
SS_warwick_dropped = SS_warwick.drop(columns=columns_to_drop)

# Display the first few rows of the modified DataFrame
display(SS_warwick_dropped.head())
````

````python
SS_warwick_dropped.dropna()
````

### Drop เฉพาะ Data ที่มี missing ใน column ที่เราสนใจ

````python
SS_warwick.dropna(subset=['Date','Gender','Age range','Self-defined ethnicity','Officer-defined ethnicity','Object of search'])
````

## เติมค่า missing ```fillna()```

### เติมด้วย dummy value (ซึ่งต้องเป็นเลขที่ไม่มีอยู่ในข้อมูล เพื่อสื่อว่าเป็นค่า missing) + เติมด้วยค่าที่น่าจะเป็นจากข้อมูลที่เรารู้

````python

set(SS_warwick['Part of a policing operation'])
````

````python
SS_warwick.fillna({'Part of a policing operation':False}, inplace=True)
````

### เติมด้วย ค่ากลาง

เติม lat long ด้วย ค่า mean

````python
SS_warwick.fillna({'Part of a policing operation':False,
                   'Latitude': SS_warwick['Latitude'].mean(),
                   'Longitude': SS_warwick['Longitude'].mean()})
````

````python
SS_warwick.fillna({'Part of a policing operation':False,
                   'Latitude': SS_warwick['Latitude'].mean(),
                   'Longitude': SS_warwick['Longitude'].mean()}).isnull().any()
````
